```python
import random

def teg():
    cont1 = 0
    cont2 = 0
    cont3 = 0
    cont4 = 0
    cont5=0
    cont6=0
    cont7=0
    cont8=0

    for i in range(10000):
        aa = [random.randint(1, 6),random.randint(1, 6),random.randint(1, 6)]
        bb = [random.randint(1, 6),random.randint(1, 6),random.randint(1, 6)]
        a = sorted(aa,reverse=True)
        b = sorted(bb, reverse=True)

        if a[0] > b[0] and a[1] > b[1] and a[2]>b[2]:
            cont1 = cont1 + 1
         # 2 ATAQUE 1 DEFENSA
        elif a[0] > b[0] and a[1] > b[1] and a[2]<=b[2]: # atacante atacante defensor
            cont2=cont2 + 1
        elif a[0] > b[0] and a[1] <= b[1] and a[2]>b[2] :
          cont3=cont3 + 1 #  atacante, defensor,atacante

        elif a[0] <= b[0] and a[1] > b[1] and a[2]>b[2]:
          cont4=cont4 + 1  # defensor atacante atacante

        #2 DEFENSA 1 ATAQUE
        elif a[0] > b[0] and a[1] <= b[1] and a[2]<=b[2] :
          cont5=cont5 + 1 # atacante defensor defensor

        elif a[0] <= b[0] and a[1] > b[1] and a[2]<=b[2]:
          cont6=cont6 + 1  # defensor atacante defensor

        elif a[0] <= b[0] and a[1] <= b[1] and a[2]>b[2]:
          cont7=cont7 + 1    # defensor defensor atacante



        else:
            cont8 = cont8 + 1
    porcentaje_cont1 = (cont1 / (i+1)) * 100
    porcentaje_cont2 = (cont2 / (i+1)) * 100
    porcentaje_cont3 = (cont3 / (i+1)) * 100
    porcentaje_cont4 = (cont4 / (i+1)) * 100
    porcentaje_cont5 = (cont5 / (i+1)) * 100
    porcentaje_cont6 = (cont6 / (i+1)) * 100
    porcentaje_cont7 = (cont7 / (i+1)) * 100
    porcentaje_cont8 = (cont8 / (i+1)) * 100


    #porcentaje = (cont1 / (cont1 + cont2)) * 100
    return cont1, cont2, cont3, cont4,cont5,cont6,cont7,cont8, porcentaje_cont1,porcentaje_cont2, porcentaje_cont3, porcentaje_cont4,porcentaje_cont5,porcentaje_cont6,porcentaje_cont7,porcentaje_cont8 #porcentaje

for i in range(6):
    cont1, cont2, cont3, cont4,cont5,cont6,cont7,cont8, porcentaje_cont1,porcentaje_cont2, porcentaje_cont3, porcentaje_cont4,porcentaje_cont5,porcentaje_cont6,porcentaje_cont7,porcentaje_cont8 = teg()
    print(f"Lanzamiento {i+1}:")
    print(f"Contador 1: {cont1}")
    print(f"2 Atacante 1 defensor : {cont2+cont3+cont4}")

    print(f"1 Atacante 2 defensor : {cont5+cont6+cont7}")
    print(f"Defensor 3-0 : {cont8}")


    print(f"3-0 atacante {porcentaje_cont1} %")
    print(f"2-1 {porcentaje_cont3+porcentaje_cont2+porcentaje_cont4} %")

    print(f"1-2  {porcentaje_cont5+porcentaje_cont6+porcentaje_cont7} %")
    print(f"0-3 {porcentaje_cont8}")
    #print(f"Porcentaje: {porcentaje:.2f}%")
    print()


```
